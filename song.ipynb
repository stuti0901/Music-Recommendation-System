```python
data.rename(columns=lambda x: x.replace(" ", "_"), inplace=True)
data
```

```python
dataf = data.drop("Singer_type",axis=1)
dataf.head()
```

```python
dataf.shape
```

```python
dataf['Song_Title_'].value_counts()
```

```python
from sklearn.feature_extraction.text import CountVectorizer
vectorizer = CountVectorizer(binary = True)
```

```python
X = vectorizer.fit_transform(dataf['Song_Title_'])
X.shape
```

```python
query1 = "har"
query2 = "koi"
list_q1 = X[:,vectorizer.vocabulary_[query1]]
list_q2 = X[:,vectorizer.vocabulary_[query2]]
```

```python
matchedsongs= 0 
for i in range(list_q1.shape[0]):
    if list_q1[i]==1 and list_q2[i]==1:
        print(dataf.iloc[i,1])
        matchedsongs = matchedsongs+1
print("Matched song"+str(matchedsongs))
```
